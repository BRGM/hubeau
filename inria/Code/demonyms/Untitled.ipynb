```python
wikidata_ids = json.load(open("Data/wikidata_ids_(1-16).json"))
exists = json.load(open("Data/wikidata_infobox_extracted_demonyms1.json"))
exists2 = json.load(open("Data/wikidata_infobox_extracted_demonyms2.json"))
treated = json.load(open("Data/treated.json"))
```

```python
len(exists), len(exists2), len(treated)
```

```python
def cond(commune):
    if commune not in exists and commune not in exists2 and commune not in treated:
        return True
    return False
```

```python
import sys
gentiles = {}
gentiles_ = {}
treated = []
for label, ids in wikidata_ids.items() :
    treated.append(label)
    print(label)
    result = set()
    if cond(label):
        for id_ in ids :
            try:
                page = wptools.page(wikibase=id_, lang='fr', silent=True).get()
                infobox = page.data["infobox"]
                if infobox is not None and "gentilé" in infobox:
                    g = infobox["gentilé"]
                    result.add(g)
                    print(result)
                    
            except KeyboardInterrupt:
                print("interrupted")
                with open("Data/wikidata_infobox_extracted_demonyms3.json", "w") as fp:
                    json.dump(gentiles,fp)
                with open("Data/treated.json", "w") as fp:
                    json.dump(gentiles_,fp)
                sys.exit()
                
            except:
                with open("Data/wikidata_infobox_extracted_demonyms3.json", "w") as fp:
                    json.dump(gentiles,fp)
                with open("Data/treated.json", "w") as fp:
                    json.dump(gentiles_,fp)
                print("page does not exist")
        
        gentiles_[label] = list(result)
        if len(result):
            gentiles[label] = list(result)
        print(gentiles_[label])

    print("_______________________________________________________________________________________")
        
with open("Data/wikidata_infobox_extracted_demonyms.json", "w") as fp:
    json.dump(gentiles,fp)
```
